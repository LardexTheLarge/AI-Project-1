```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
columns_to_select = ['Country', 'Project type', 'Announcement','Project Status',\
                     'Project phase','Announced capacity (low) (Mt CO2/yr)','Sector',\
                        'Announced capacity (high) (Mt CO2/yr)']

df = pd.read_excel('data/IEA CCUS Projects Database 2023.xlsx',\
                   sheet_name = 'CCUS Projects Database', usecols=columns_to_select)

df.set_index('Announcement',inplace=True)
# Display the first few rows of the dataset
df.head()
```

```python
# Basic information about the dataset
df.info()
```

```python
#view global project status.
df['Project Status'].value_counts()
```

```python
#evaluate data set by country and project status
pd.crosstab(index=df['Country'], columns=df['Project Status']).\
    sort_values(by = 'Operational', ascending= False)
```
